```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('final.csv')
```

```python
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
data.duplicated().any()
```

```python
data['Price'].unique()
```

```python
data['Price'] = data['Price'].str.strip('$')
```

```python
# Replace 'Price unavailable' with None for further analysis
data.replace({'Price unavailable': None}, inplace=True)
```

```python
for i in range (652):
    if data['Price'].loc[i] is None:
        continue
    else:
        data['Price'].loc[i] = int(data['Price'].loc[i].replace(',',''))
```

```python
data['Year'] = np.nan
```

```python
for i in range(652):
    data['Year'].loc[i] = str(data['Title'].loc[i])[0:4]
```

```python
data.loc[data['Miles'].str.contains('K'), 'Miles'] = data.Miles.str.strip('K').astype(float)*1000
```

```python
data.Score.unique()
```

```python
for i in range(652):
    if data.Score.loc[i].isnumeric() == False:
        data.Score.loc[i] = np.nan
```

```python
for i in range(652):
    if data.Drive_Type.loc[i] == 'Automatic\nT':
        data.Drive_Type.loc[i] = 'Automatic'
```

```python
for i in range(652):
    if data.Drive_Type.loc[i] == 'Automatic':
        data.Prior_Use.loc[i] = data.Color.loc[i]
        data.Color.loc[i] = data.Transmission.loc[i]
        data.Transmission.loc[i] = 'Automatic'
        data.Drive_Type.loc[i] = np.nan
```
